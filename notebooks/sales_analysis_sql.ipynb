```python
import pandas as pd

# Load the dataset
file_path = "/content/train.csv"
df = pd.read_csv(file_path)


df.info(), df.head()
```

```python

df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d/%m/%Y')

# Fill missing postal codes
df['Postal Code'].fillna(method='ffill', inplace=True)

# Convert 'Postal Code' to integer type
df['Postal Code'] = df['Postal Code'].astype(int)

df.info(), df.head()
```

```python
import sqlite3

conn = sqlite3.connect("/content/sales_data.db")
cursor = conn.cursor()

df.to_sql("sales", conn, if_exists="replace", index=False)

cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = cursor.fetchall()
tables
```

```python
query_total_sales = "SELECT SUM(Sales) AS Total_Revenue FROM sales;"
total_sales = pd.read_sql(query_total_sales, conn)

display(total_sales)

```

```python
#Sales by Region
query_sales_by_region = """
SELECT Region, SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Region
ORDER BY Total_Sales DESC;
"""
sales_by_region = pd.read_sql(query_sales_by_region, conn)

display(sales_by_region)
```

```python
#Sales by Product Category
query_sales_by_category = """
SELECT Category, SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Category
ORDER BY Total_Sales DESC;
"""
sales_by_category = pd.read_sql(query_sales_by_category, conn)

display(sales_by_category)
```

```python
#Sales by Customer Segment
query_sales_by_segment = """
SELECT Segment, SUM(Sales) AS Total_Sales
FROM sales
GROUP BY Segment
ORDER BY Total_Sales DESC;
"""
sales_by_segment = pd.read_sql(query_sales_by_segment, conn)

display(sales_by_segment)
```

```python

import numpy as np

df_cleaned = df.copy()


df_cleaned.dropna(inplace=True)

# Removing duplicates
df_cleaned.drop_duplicates(inplace=True)

# Standardizing categorical columns
categorical_cols = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub-Category']
df_cleaned[categorical_cols] = df_cleaned[categorical_cols].apply(lambda x: x.str.strip().str.lower())

df_cleaned['Order Year'] = df_cleaned['Order Date'].dt.year
df_cleaned['Order Month'] = df_cleaned['Order Date'].dt.month
df_cleaned['Order Day'] = df_cleaned['Order Date'].dt.day


df_cleaned['Log Sales'] = np.log1p(df_cleaned['Sales'])

display(df_cleaned)
```
